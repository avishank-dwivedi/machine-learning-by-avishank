```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset  = pd.read_csv("Data.csv")
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x = "Salary" ,data=dataset)
plt.show()
```

```python
sns.distplot(dataset["Salary"])
```

```python
dataset.shape
```

```python
q1 = dataset["Salary"].quantile(0.25)
q3 = dataset["Salary"].quantile(0.75)
```

```python
q1
```

```python
q3
```

```python
IQR = q3 -q1
```

```python
min_range = q1 -(1.5*IQR)
max_range = q3 +(1.5*IQR)
min_range , max_range
```

```python
nw_dataset = dataset[dataset["Salary"]<=max_range]
```

```python
nw_dataset.shape
```

```python
sns.boxplot(x = "Salary" ,data=nw_dataset)
plt.show()
```
