```python
min_range = dataset["YearsInCurrentRole"].mean()-(3*dataset["YearsInCurrentRole"].std())
max_range = dataset["YearsInCurrentRole"].mean()+(3*dataset["YearsInCurrentRole"].std())
```

```python
min_range,max_range
```

```python
new_data = dataset[dataset["YearsInCurrentRole"]<max_range]
```

```python
new_data.shape
```

```python

sns.boxplot(x = "YearsInCurrentRole" ,data=new_data)
plt.show()
```

```python
z_score = (dataset["YearsInCurrentRole"] - dataset["YearsInCurrentRole"].mean())/(dataset["YearsInCurrentRole"].std())
```

```python
dataset["z_score"] = z_score
```
