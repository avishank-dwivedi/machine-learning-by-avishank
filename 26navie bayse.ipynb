```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv("placement1.csv")
dataset.head(3)
```

```python
dataset.isnull().count()
```

```python
#sns.kdeplot(data=dataset["cgpa"])
#plt.show()
```

```python
sns.scatterplot(x="cgpa", y="resume_score", data=dataset, hue="placed")
plt.show()
```

```python
X = dataset.iloc[:,:-1]
Y = dataset["placed"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
gnb.score(X_test , y_test)*100 , gnb.score(X_train , y_train)*100
```

```python
plot_decision_regions(X=X.to_numpy(), y=Y.to_numpy(), clf=gnb)
plt.show()
```

```python
gnb.predict([[6.17 , 5.17]])
```

```python
mnb = MultinomialNB()
mnb.fit(X_train ,y_train)
```

```python
mnb.score(X_train,y_train)*100 , mnb.score(X_test,y_test)*100
```

```python
plot_decision_regions(X=X.to_numpy(), y=Y.to_numpy(), clf=mnb)
plt.show()
```

```python
bnb = BernoulliNB()
bnb.fit(X_train ,y_train)
```

```python
bnb.score(X_train , y_train)*100 , bnb.score(X_test , y_test)*100
```

```python
plot_decision_regions(X=X.to_numpy(), y=Y.to_numpy(), clf=bnb)
plt.show()
```
