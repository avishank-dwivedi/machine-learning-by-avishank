```python
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
import numpy as  np
```

```python
dataset = pd.read_csv("Data.csv")
```

```python
dataset.head(3)
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["Age"])
plt.show()
```

```python
q1 = dataset["Age"].quantile(0.25)
q3 = dataset["Age"].quantile(0.75)
iqr = q3-q1
```

```python
minn = q1-(1.5*iqr)
maxx = q3+(1.5*iqr)
```

```python
minn ,maxx
```

```python
data =dataset[dataset["Age"]<=maxx]
```

```python
sns.distplot(data["Age"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft =FunctionTransformer(func = np.log1p )
```

```python
ft.fit(data[["Age"]])
```

```python
data["Age_tr"] = ft.transform(data[["Age"]])
```

```python
plt.subplot(1,2,1)
sns.distplot(dataset["Age"])
plt.title("before")

plt.subplot(1,2,2)
sns.distplot(data["Age_tr"])
plt.title("after")
plt.show()
```
