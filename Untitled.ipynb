```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\Avishank Dwivedi\OneDrive\Desktop\HR_Analytics.csv")
```

```python
dataset.head(5)
```

```python
dataset.isnull().sum()
```

```python
dataset["YearsWithCurrManager"].fillna(dataset["YearsWithCurrManager"].mean(),inplace = True)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["TotalWorkingYears"])
plt.show
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(dataset[["TotalWorkingYears"]])
```

```python
dataset["TotalWorkingYears_ss"] = pd.DataFrame(ss.transform(dataset[["TotalWorkingYears"]]),columns=["X"])
```

```python
dataset.head(5)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["TotalWorkingYears"])

plt.subplot(1,2,2)
sns.distplot(dataset["TotalWorkingYears_ss"])
plt.title("After")

plt.show()
```
