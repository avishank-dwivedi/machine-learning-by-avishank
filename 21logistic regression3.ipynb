```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# logistic regressin in polymorpism

```python
plt.figure(figsize=(3,4))
dataset = pd.read_csv("ushape.csv")
dataset.head(3)
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x="3.159499363321345566e-02",y="9.869877579082642072e-01",data=dataset , hue="0.000000000000000000e+00")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset["0.000000000000000000e+00"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf = PolynomialFeatures(degree=3)
pf.fit(x)
x = pd.DataFrame(pf.transform(x))
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test , y_test)*100
```
